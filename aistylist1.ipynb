```python
#filling missing values
object_columns = df.select_dtypes(include=['object']).columns
```

```python
print("\nMissing values in object columns before processing:")
print(df[object_columns].isnull().sum())
```

```python
for col in object_columns:
    most_frequent = df[col].mode()[0]
    df[col].fillna(most_frequent, inplace=True)
```

```python
print("\nMissing values in object columns after processing:")
print(df[object_columns].isnull().sum())
```

```python
print("\nCleaned DataFrame:")
print(df.info())
```

```python
df.isnull().sum()
```

```python
if 'year' in df.columns:
    most_frequent_year = df['year'].mode()[0]
    df['year'].fillna(most_frequent_year, inplace=True)
```
